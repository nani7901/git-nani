```python
# Loop through each column and print the column name, unique values, and data type
for column in df.columns:
    print(f"Column: {column}")
    print(f"Data Type: {df[column].dtype}")
    print(f"Unique Values: {df[column].unique()}")
    print("\n")
```


*   The 'DOL' column is not in datetime format and contains the value 'present'.

*  The 'JobCity', 'Specialization', and 'Board of Examination' columns contain several redundant values.

*   The 'GraduationYear' column contains entries with the value '0'.

*   The 'JobCity' column contains entries with the value '-1'.

*   All the score value columns contain entries with the value '-1'.










```python
# changing the dtype of 'DOL' column and replacing the value 'present' with current time.
# Replace 'present' with NaT (Not a Time) first, then convert to datetime
df['DOL'] = df['DOL'].replace('present', pd.NaT)  # Replace 'present' with NaT
df['DOL'] = pd.to_datetime(df['DOL'])
df['DOL'].fillna(pd.Timestamp.now(), inplace=True)  # Replace NaT with current time
```

```python
# Check the data type of the 'DOL' column and values.
print(df['DOL'])
```

```python
# Calculate Tenure in months
df['DOJ'] = pd.to_datetime(df['DOJ'])
df['DOL'] = pd.to_datetime(df['DOL'])
# Calculate the difference in days and then divide by the average number of days in a month (30.4375)
df['Tenure'] = (df['DOL'] - df['DOJ']).dt.days / 30.4375
df['Tenure'] = df['Tenure'].astype(int)

# Display unique Tenure values
df['Tenure'].unique()
```

The 'Tenure' column contains some negative values, which may indicate that certain individuals rejected the job offer.


```python
# CREATING AGE COLUMN USING 'DOB' COLUMN
df['DOB'] = pd.to_datetime(df['DOB'])
current_date = datetime.now()

# Calculate the age in years using relativedelta and a lambda function
df['Age'] = df['DOB'].apply(lambda dob: relativedelta(current_date, dob).years)

# Display the first few rows to confirm the 'Age' column is created correctly
print(df[['DOB', 'Age']].head())
```

**Why Create an Age Column ?**
  Creating an age column from the â€˜DOBâ€™ data helps in demographic analysis, correlation studies, predictive modeling, and informed decision-making.

```python

from scipy.stats import chi2_contingency
contingency_table = pd.crosstab(df['Gender'], df['Specialization'])
print("Contingency Table:")
print(contingency_table)
chi2, p, _, _ = chi2_contingency(contingency_table)
print("\nChi-squared statistic:", chi2)
print("P-value:", p)
alpha = 0.05 # Set the significance level
print("\nSignificance level:", alpha)
if p < alpha:
    print("There is a significant relationship between Gender and Specialization.")
else:
    print("There is no significant relationship between Gender and Specialization.")
```
